```python
##Importing Selenium
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import time
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
url  = "https://dir.indiamart.com/search.mp?ss=camera&mcatid=22&catid=204&prdsrc=1&res=RC5"
```

```python
service = Service(executable_path="./chromedriver")
```

```python
driver = webdriver.Chrome(service=service)
```

```python
driver.get(url)
```

```python
#sleep 
time.sleep(10)
```

```python
#scrolling to bottom of the page
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
time.sleep(10)
```

```python
page_c=1
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
while True:
    try:
        button_picker=driver.find_element(By.CSS_SELECTOR,".dib.fs15.br5.clr2.fwb.bg3.cp")
    except Exception as e:
        print(e)
        break
    print(page_c)
    button_picker.click()
    time.sleep(5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")    
    page_c+=1
   
```

```python
# Parsing the Html
r=requests.get(url)
r.content
htmlContent=r.content
# print(htmlContent)
```

```python
#Soup element
soup= BeautifulSoup(driver.page_source,'html.parser')
# print(soup)
```

```python
# Getting <a> tags in the soup
Tot_no_products= soup.find_all("a", class_="prd-name")
len(Tot_no_products)
```

```python
#Getting all of the section cards on the page
final_content= soup.find_all(class_="lst_cl prd-card fww brs5 pr bg1 prd-card-mtpl")
len(final_content)
```

```python
# Getting first section/card
final_content[0]
```

```python
for sec in final_content[0]:
    print(len(sec))
```

```python
#Extracting all of the data collected from each card

Product_Location=[]
Product_Vendor=[]
Contact=[]
Product_Name=[]
Final_list=[]
empty_dict={}

for sec in final_content:
    
    target_product_name= sec.select(".prd-name")
    P_name_data=[tag.text for tag in target_product_name]
    print(P_name_data)
    Product_Name.extend(P_name_data)
    
    target_location_name= sec.select(".elps.elps1")
    product_location_data= [tag.text for tag in target_location_name]
    print(product_location_data*len(P_name_data))
    Product_Location.extend(product_location_data)
    
    target_vendors_name= sec.select(".clr3.fs12.fwn.rsrc")
    vendor_name_data = [attribute.find("span").text for attribute in target_vendors_name]
    print(vendor_name_data*len(P_name_data))
    Product_Vendor.extend(vendor_name_data)

    target_contact_no= sec.select(".pns_h.duet.fwb")
    contacts_data= [tag.text for tag in target_contact_no]
    print(contacts_data*len(P_name_data))
    Contact.extend(contacts_data)
    
    Product_data={ 
            'Pvendor'         :  vendor_name_data ,  
            'Plocation'       :  product_location_data,
            'Pcontact'       : contacts_data,
            'Pname'          : P_name_data
        
    }
    Final_list.append(Product_data)
    
    
    
    
    
    
```
